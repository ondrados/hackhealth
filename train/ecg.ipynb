```python
import re
import os
import glob
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle("data/train.pkl")
```

```python
df.ecg_path.iloc[0]
```

```python
sig = pd.read_csv(df.ecg_path.iloc[0])
sig
```

```python
leads = [" V1", " V3", " V6"]
fig, axes = plt.subplots(nrows=3,ncols=1,figsize=(30,15))
for lead in range(len(leads)):
    axes[lead].plot(sig[leads[lead]],label=leads[lead])
    axes[lead].set_xlabel('samples')
    axes[lead].set_title(leads[lead])
plt.show()
```
