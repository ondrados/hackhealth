```python
import re
import os
import glob
import pickle
import numpy as np
import pandas as pd
```

```python
splits = ["kkchb_hosp", "kd_hosp", "assorted"]
data_folder = f"/Users/ondra/dev/hackhealth/data/{splits}_csv"
csv_paths = glob.glob(data_folder + "/*csv")
ids = [os.path.basename(path).split(".")[0] for path in csv_paths]
```

```python
dfs = []
for split in splits:
    dfs.append(pd.read_pickle(f"data/{split}_df.pkl"))
df = pd.concat(dfs)
```

```python
len(df.id.unique())
```

```python
df_as = pd.read_csv("/Users/ondra/dev/hackhealth/data/as_operations.csv", sep=";")
```

```python
df["as"] = None

df.id = df.id.apply(lambda x: x.replace("/", ""))
#df.id = df.id.apply(lambda x: x.replace("n", ""))
df_as.id = df_as.id.apply(lambda x: x.replace("/", ""))
df_as.id = df_as.id.apply(lambda x: re.sub("[^0-9]", "", x))

df_ids = list(df.id)
df_as_ids = list(df_as.id)

for as_id in df_as_ids:
    df["as"].loc[df.id == as_id] = 1
```

```python
as_count = len(df.loc[df["as"] == 1])
print(f"AS: {as_count}")
```
